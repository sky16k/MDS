### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```
import sympy as sym
import numpy as np
```

```
### YOUR CODE HERE

x = Symbol('x')
y = x**5+4*sin(2*x)+cos(3*x+3)
```

```
y_der1 = y.diff(x)
# print(
y_der1
# )
```

```
y_der2 = y_der1.diff(x)
y_der2
```

```
print('Значение первой производной в точке x = 1:')
y_der1.subs(x,1).evalf()
```

```
print('Значение второй производной в точке x = 1:')
y_der2.subs(x,1).evalf()
```
